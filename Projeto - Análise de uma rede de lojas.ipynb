### União das abas por colunas.
<br>
<li><strong>Aba e Aba1</strong> tem a coluna <strong>ProductID</strong> e será por ela que nós uniremos todos os dados das duas abas</li>
<li><strong>Aba e Aba2</strong> tem a coluna <strong>ClientID</strong> e será por ela que nós uniremos todos os dados das duas abas</li>
<li><strong>Aba e Aba3</strong> tem a coluna <strong>StoreID</strong> e será por ela que nós uniremos todos os dados das duas abas</li>
<li><strong>Aba3 e Aba2</strong> tem as colunas <strong>StoreID e State</strong>,  e será por ela que nós uniremos todos os dados das duas abas</li>

```python
uniao1 = pd.merge(aba, aba1, how = 'outer',on='ProductID') 
uniao2 = pd.merge(aba, aba2, how = 'outer',on='ClientID')
uniao3 = pd.merge(aba, aba3, how = 'outer',on='StoreID')
uniao4 = pd.merge(aba3[['StoreID',  'State']],aba2, how = 'outer',on='State')
uniao_geral = pd.merge(uniao1, aba2, how = 'outer',on='ClientID')
```

### Gráficos de comparação entre as lojas e seus crescimentos e acelerações.

```python
def grafico_comparacao(loja,ano,ano1,ano2,x,y1,y2,y3):
    analise_cl = pd.DataFrame(dados)
    analise_cl['dia'] = analise_cl['Date'].dt.day
    analise = analise_cl.groupby(['StoreID','dia','mes','ano'])[['Quantity']].agg(['sum'])
    analise.reset_index(inplace=True)
    analise.rename(columns = {'sum':'',}, inplace = True)
    analise = pd.DataFrame(analise)
    analise.columns = analise.columns.droplevel(1)
    analise = analise.sort_values(by=['ano','StoreID'])
    analise['aumento'] = analise['Quantity'].diff()
    analise['aceleracao'] = analise['aumento'].diff()
    
    df = analise.query(f'StoreID == {loja} & ano == {ano}')
    df1 = analise.query(f'StoreID == {loja} & ano == {ano1}')
    df2 = analise.query(f'StoreID == {loja} & ano == {ano2}')
    
    mes = analise.mes.unique()
    titulo = str(f'Análise de vendas da loja {loja} nos anos de {ano}, {ano1} e {ano2}')

    sns.set_theme(style="ticks")
    plt.figure(figsize=(12,7)) 
    
    palette = sns.husl_palette(1, h=.7)
    palette1 = sns.husl_palette(1, h=.5)
    palette2 = sns.husl_palette(1, h=.9)
    
    grafico = plt.subplot(3,1,1)
    grafico.set_title(titulo,fontsize=18,loc='center')
    sns.lineplot(data = df , x=x, y=y1,hue="ano",palette =palette )
    sns.lineplot(data = df1, x=x, y=y1,hue="ano",palette =palette1)
    sns.lineplot(data = df2, x=x, y=y1,hue="ano",palette =palette2)
    plt.xticks(ticks=mes, size='small', color='black')
    plt.grid(axis='x')
    plt.legend(title='Ano',bbox_to_anchor=(0.6, 0.5, 0.5, 0.5))
    
    plt.subplot(3,1,2)
    sns.lineplot(data = df , x=x, y=y2,hue="ano",palette =palette )
    sns.lineplot(data = df1, x=x, y=y2,hue="ano",palette =palette1)
    sns.lineplot(data = df2, x=x, y=y2,hue="ano",palette =palette2)
    plt.xticks(ticks=mes, size='small', color='black')
    plt.grid(axis='x')
    plt.legend(title='Ano',bbox_to_anchor=(0.6, 0.5, 0.5, 0.5))
    
    plt.subplot(3,1,3)
    sns.lineplot(data = df, x= x, y = y3,hue="ano",palette =palette )
    sns.lineplot(data= df1, x= x, y = y3,hue="ano",palette =palette1)
    sns.lineplot(data= df2, x= x, y = y3,hue="ano",palette =palette2)
    plt.xticks(ticks=mes, size='small', color='black')
    plt.grid(axis='x')
    plt.legend(title='Ano',bbox_to_anchor=(0.6, 0.5, 0.5, 0.5))

#     sns.despine(right=False,bottom=False,top=False)
    plt.tight_layout()
    grafico=grafico
    
grafico_comparacao(1, 2018,2019,2020,'mes','Quantity','aumento','aceleracao')
```
